```python
"""
Problem Statement: 16
You are provided with a transactional market basket dataset 
(market_basket_dataset.csv), where each transaction contains a set of items 
purchased together by a customer. Develop a program to demonstrate 
Association Rule Mining using the Apriori Algorithm. 
The program should perform the following tasks: 
 
i. Read the transactional dataset from a CSV file. 
ii. Convert the dataset into a suitable format for association rule mining. 
iii. Apply the Apriori algorithm to generate frequent itemsets based on a 
minimum support threshold. 
iv. Generate association rules from the frequent itemsets using minimum 
confidence. 
v. Display the discovered frequent itemsets and association rules along with 
their support, confidence values. 
"""
```

```python
#Importing libraries
import pandas as pd
import random
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
# Creating dataset
items = ["Milk", "Bread", "Butter", "Eggs", "Cheese", "Juice", "Apples", "Banana"]

transactions = []

for _ in range(100):
    transaction = random.sample(items, random.randint(2, 5))
    transactions.append(transaction)

df = pd.DataFrame({"Transaction": transactions})

df.to_csv("market_basket_dataset.csv", index=False)

print("Dataset created!")
```

```python
# i. Read the transactional dataset
df = pd.read_csv('market_basket_dataset.csv') 
```

```python
# ii. Convert the dataset into a suitable format (One-Hot Encoding)
# The Apriori algorithm requires a boolean/binary matrix
transactions = df["Transaction"].apply(eval).tolist()
te = TransactionEncoder()
te_array = te.fit(transactions).transform(transactions)
df_encoded = pd.DataFrame(te_array, columns=te.columns_)
```

```python
# iii. Apply Apriori to generate frequent itemsets
frequent_items = apriori(df_encoded, min_support=0.2, use_colnames=True) 
```

```python
# iv. Generate association rules using minimum confidence
rules = association_rules(frequent_items, metric="confidence", min_threshold=0.6)
```

```python
# v. Display the discovered frequent itemsets
print("\nFrequent Itemsets:")
print(frequent_items)

rules = association_rules(frequent_items, metric="confidence", min_threshold=0.6)

print("\nAssociation Rules:")
print(rules[["antecedents", "consequents", "support", "confidence"]])
```
