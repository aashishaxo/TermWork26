```python
"""
Problem Statement: 4
Design and develop programs to implement the Depth-First Search (DFS) 
algorithm for graph traversal in Artificial Intelligence with the following 
specifications: 
i. Construct a static graph consisting of 6 nodes and 10 edges, 
where the graph is stored using a dictionary representation. 
The graph is given as: A- B, A-C, B-D, B-E, C-D, C-F, D-E, 
D-F, E-F, B-F.  
ii. Create another graph where the number of nodes and number 
of edges are provided by the user at runtime. 
iii. Accept the initial node and the goal node from the user. 
iv. Display the path from the initial node to the goal node for both 
the cases. 
"""
```

```python
#Graph for Static DFS 
graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E', 'F'],
    'C': ['A', 'D', 'F'],
    'D': ['B', 'C', 'E', 'F'],
    'E': ['B', 'D', 'F'],
    'F': ['C', 'D', 'E', 'B']
}

# DFS For Static Graph
def DFS(graph, node, visited, goal, path):
    visited.add(node)
    path.append(node)

    if node == goal:
        return True

    for n in graph[node]:
        if n not in visited:
            if DFS(graph, n, visited, goal, path):
                return True

    # Backtracking step
    path.pop()
    return False


visited = set()
path = []

start = input("Enter start point: ")
goal = input("Enter Goal point: ")

if DFS(graph, start, visited, goal, path):
    print("\nSuccess")
    print("Path:", " -> ".join(path))
else:
    print("\nFailed")
```

```python
#Dynamically taking graph input from the user
graph = {}
n = int(input("Enter number of nodes:"))
for i in range(n):
    node = input("Enter Node:")
    neighbour = input("Enter Node Neighbours:").split()
    graph[node] = neighbour
print("Graph:",graph)
start = input("Enter start point:")
goal = input("Enter goal point:")

visited = set()
path = []

#Calling the DFS function
if DFS(graph, start, visited, goal, path):
    print("\nSuccess")
    print("Path:", " -> ".join(path))
else:
    print("\nFailed")
```
