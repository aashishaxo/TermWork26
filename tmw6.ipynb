```python
"""
Problem Statement: 6
Construct a program to implement the A* search algorithm using heuristic 
values for a given graph. The graph is stored using a dictionary 
representation. The graph consists of the following nodes and edge costs: 
Path          Edge Cost 
A-B           1 
A-C           3 
B-D           3 
B-E           6 
C-D           1 
D-E           1 
 
Node A B C D E 
h(n) 7 6 4 2 0 
 
Accept the initial node from the user and node E as goal node. Apply the 
A* algorithm to: 
i. Display the order of node expansion 
ii. Display the optimal path 
iii. Display the total path cost 
"""
```

```python
import heapq

# Graph representation (adjacency list with cost)
graph = {
    'A': {'B': 1, 'C': 3},
    'B': {'D': 3, 'E': 6},
    'C': {'D': 1},
    'D': {'E': 1},
    'E': {}
}

# Heuristic values
heuristic = {
    'A': 7,
    'B': 6,
    'C': 4,
    'D': 2,
    'E': 0
}

# A* Algorithm
def A_star(graph, heuristic, start, goal):
    open_list = [start]
    closed_list = set()

    g = {start: 0}   # cost from start
    parent = {start: None}

    expansion_order = []

    while open_list:
        # find node with lowest f(n) = g(n) + h(n)
        current = open_list[0]
        for node in open_list:
            if g[node] + heuristic[node] < g[current] + heuristic[current]:
                current = node

        open_list.remove(current)
        closed_list.add(current)
        expansion_order.append(current)

        # goal reached
        if current == goal:
            path = []
            while current is not None:
                path.append(current)
                current = parent[current]
            path.reverse()
            return expansion_order, path, g[goal]

        # explore neighbors
        for neighbor in graph[current]:
            cost = g[current] + graph[current][neighbor]

            if neighbor not in open_list and neighbor not in closed_list:
                open_list.append(neighbor)
                parent[neighbor] = current
                g[neighbor] = cost

            else:
                if cost < g.get(neighbor, float('inf')):
                    g[neighbor] = cost
                    parent[neighbor] = current
                    if neighbor in closed_list:
                        closed_list.remove(neighbor)
                        open_list.append(neighbor)

    return None, None, None
    
# User Input Start Node and Processing
start = input("Enter start node: ")
goal = 'E'

expansion, path, cost = A_star(graph, heuristic, start, goal)

print("Expansion Order:", " -> ".join(expansion))
print("Optimal Path:", " -> ".join(path))
print("Total Cost:", cost) 
```
