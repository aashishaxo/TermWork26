```python
#Importing Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = {
    "height_cm": [160, 172, 168, 180, 155, 165, 170, 178, 162, 175],
    "weight_kg": [55, 70, 62, 80, 50, 60, 68, 75, 58, 72],
    "gender": ["Female", "Male", None, "Male", "Female", None, "Male", "Male", "Female", "Male"]
}
df = pd.DataFrame(data) 
df
```

```python
#i.) Compute BMI using dataframe operations. 
df["height_m"] = df["height_cm"] / 100
df["BMI"] = df["weight_kg"] / (df["height_m"] ** 2)
print("\nDataset with BMI:\n", df) 
```

```python
#ii.) Save the dataset into CSV file. 
df.to_csv("bmi_dataset.csv", index=False) 
```

```python
#iii.) Import the csv file and load it into a dataframe. 
df2 = pd.read_csv("bmi_dataset.csv")
df2 
```

```python
#iv.) Find shape and size of dataframe. 
print("\nShape:", df2.shape)
print("Size:", df2.size) 
```

```python
#v.) Find the count of null values in Gender column. 
print("\nNull values in Gender:", df2["gender"].isnull().sum()) 
```

```python
#vi.) Remove null from gender column. 
df2 = df2.dropna(subset=["gender"])
print("\nAfter removing nulls:\n", df2) 
```

```python
#vii.) Display the structural and statistical information of the dataset. 
print("\nInfo:")
print(df2.info()) 
print("\nStatistics:")
print(df2.describe()) 
```

```python
#viii.) Create a histogram on gender column. 
df2["gender"].value_counts().plot(kind='bar')
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show() 
```

```python
#ix.) Draw a suitable plot on the data and deduce some inference. 
plt.scatter(df2["height_cm"], df2["weight_kg"])
plt.title("Height vs Weight")
plt.xlabel("Height (cm)")
plt.ylabel("Weight (kg)")
plt.show()
print("\nInference: Generally, weight increases with height (positive correlation).") 
```

```python
#x.) Perform aggregate functions on height, weight and BMI column.
print("\nAggregate values:")
print("Height Mean:", df2["height_cm"].mean())
print("Weight Mean:", df2["weight_kg"].mean())
print("BMI Mean:", df2["BMI"].mean())

print("Max BMI:", df2["BMI"].max())
print("Min BMI:", df2["BMI"].min()) 
```

```python
#xi.) Find the entries where BMI is less than 20 
print("\nEntries with BMI < 20:\n", df2[df2["BMI"] < 20]) 
```
